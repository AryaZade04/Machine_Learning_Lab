```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the Excel file
#file_path = "Houseprice.csv.xlsx"
df = pd.read_excel("Houseprice.csv.xlsx" )
```

```python
df.head()
```

```python
df.describe()
```

```python
X = df[['Area (sq. ft.)']]  # Independent
y = df['Price (₹ Lakhs)']   # Dependent
```

```python
# Create and train the model
model = LinearRegression()
model.fit(X, y)
```

```python
# Make predictions
y_pred = model.predict(X)
```

```python
# Get regression coefficients
slope = model.coef_[0]
intercept = model.intercept_
```

```python
# Compute R-squared value
r_squared = r2_score(y, y_pred)
```

```python
# Print results
print(f"Regression Equation: Price = {slope:.4f} * Area - {abs(intercept):.2f}")
print(f"R-squared Score: {r_squared:.4f}")
```

```python

# Plotting
plt.figure(figsize=(10, 6))
sns.scatterplot(x=X['Area (sq. ft.)'], y=y, color='blue', label='Actual Data')
sns.lineplot(x=X['Area (sq. ft.)'], y=y_pred, color='red', label='Regression Line')
plt.title('Linear Regression: Area vs Price')
plt.xlabel('Area (sq. ft.)')
plt.ylabel('Price (₹ Lakhs)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
