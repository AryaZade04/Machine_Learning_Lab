```python
"""
simple_linear_regression.py

Simple Linear Regression Example:
Predicting test scores from hours studied.
"""

import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
# -------------------------
# 1) Data
# -------------------------
X = np.array([[1], [2], [3], [4], [5]])   # hours studied
y = np.array([20, 40, 50, 65, 80])        # test scores

print("X (hours studied):\n", X.ravel())
print("y (scores):\n", y)
```

```python
# -------------------------
# 2) Train model
# -------------------------
model = LinearRegression()
model.fit(X, y)

slope = model.coef_[0]
intercept = model.intercept_

print("\nModel parameters:")
print(f"Slope (m): {slope:.2f}")
print(f"Intercept (b): {intercept:.2f}")
print(f"Equation: y = {slope:.2f} * x + {intercept:.2f}")
```

```python
# -------------------------
# 3) Predictions
# -------------------------
y_pred = model.predict(X)

print("\nPredictions vs Actual:")
for xi, yi, yp in zip(X.ravel(), y, y_pred):
    print(f"Hours={xi}, Actual Score={yi}, Predicted Score={yp:.2f}")
```

```python
# -------------------------
# 4) Evaluation metrics
# -------------------------
mae = mean_absolute_error(y, y_pred)
mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)

print("\nEvaluation Metrics:")
print(f"MAE  = {mae:.2f}")
print(f"MSE  = {mse:.2f}")
print(f"R^2  = {r2:.4f}")
```

```python

# -------------------------
# 5) Visualization
# -------------------------
plt.scatter(X, y, color="blue", label="Actual data")
plt.plot(X, y_pred, color="red", linewidth=2, label="Regression line")
plt.xlabel("Hours Studied")
plt.ylabel("Test Score")
plt.title("Simple Linear Regression: Hours vs Score")
plt.legend()
plt.grid(True)
plt.show()
```
